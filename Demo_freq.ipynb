```python
import numpy as np
import pandas as pd
import kavutil as ku
```

Zunächst werden Daten simuliert. Die Variable Bildung erhält theoretisch 3% missings (mit diesem random seed sind es 2,2%)

```python
level_bildung = ["Hauptschule", "Realschule", "Gymnasium", "Studium"]
level_geschlecht = ["männlich", "weiblich"]

np.random.seed(90491)
bildung = np.random.choice(level_bildung, 500, p=[0.5, 0.25, 0.20, 0.05])
geschlecht = np.random.choice(level_geschlecht, 500)
einkommen = np.round_(np.random.normal(2500, 2000, 500), decimals=2)
missing = np.random.binomial(1, 0.03, 500)

daten = pd.DataFrame(dict(Bildung = bildung, Geschlecht = geschlecht, Einkommen = einkommen))
daten.loc[missing==1, "Bildung"] = np.NaN

daten.head()
```

Häufigkeitstabelle für Daten ohne Missings

```python
ku.freq(daten, "Geschlecht")
```

Häufigkeitstabelle für Daten mit Missings

```python
ku.freq(daten, "Bildung")
```

Häufigkeitstabelle für Daten ohne Missings, aber mit Option nomiss=True

```python
ku.freq(daten, "Bildung", nomiss=True)
```
